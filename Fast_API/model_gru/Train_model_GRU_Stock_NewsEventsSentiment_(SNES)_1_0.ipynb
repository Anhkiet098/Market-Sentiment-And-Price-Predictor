```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Đọc dữ liệu
data = pd.read_csv('/content/drive/MyDrive//data/data.csv')

# Xử lý dữ liệu trống
data.fillna(0, inplace=True)

# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume',
      'News - Positive Sentiment', 'News - Negative Sentiment']] = scaler.fit_transform(
    data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume',
          'News - Positive Sentiment', 'News - Negative Sentiment']])

# Xử lý mã cổ phiếu (chuyển đổi thành số hoặc one-hot encoding nếu cần)
symbol_mapping = {symbol: idx for idx, symbol in enumerate(data['Symbol'].unique())}
data['Symbol'] = data['Symbol'].map(symbol_mapping)

# Tạo tập dữ liệu huấn luyện
X = data[['Open', 'High', 'Low', 'Adj Close', 'Volume', 'Symbol',
          'News - Positive Sentiment', 'News - Negative Sentiment']]
y = data['Close']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Chuyển đổi dữ liệu đầu vào thành định dạng 3D cho GRU
X_train = np.reshape(X_train.values, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test.values, (X_test.shape[0], 1, X_test.shape[1]))

# Xây dựng mô hình GRU
model = Sequential()
model.add(GRU(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(GRU(units=50))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Lưu mô hình tốt nhất
checkpoint = ModelCheckpoint('/content/drive/MyDrive/model5/gru_model.keras', save_best_only=True, monitor='val_loss', mode='min')
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[checkpoint])

# Dự đoán trên tập test
y_pred = model.predict(X_test)

# Đánh giá mô hình
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'R2 Score: {r2}')

# Biểu đồ so sánh giữa thực tế và dự đoán
plt.figure(figsize=(14, 5))
plt.plot(y_test.values, label='Giá thực tế')
plt.plot(y_pred, label='Giá dự đoán', color='red')
plt.legend()
plt.title('So sánh giá thực tế và dự đoán')
plt.show()
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime, timedelta
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt


def predict_next_7_days(symbol, model_path):
    """
    Predict stock prices for the next 7 days using trained GRU model
    """
    # Load the model
    model = load_model(model_path)

    # Get historical data from yfinance
    end_date = datetime.now()
    start_date = end_date - timedelta(days=30)
    stock_data = yf.download(symbol, start=start_date, end=end_date)

    # Create scaler and fit on historical data
    scaler = MinMaxScaler()
    stock_data_scaled = scaler.fit_transform(
        stock_data[[d]]
    )

    # Get the latest data
    latest_data = stock_data_scaled[-1]

    # Initialize predictions list
    predictions = []
    dates = []

    # Prepare initial input
    current_input = np.zeros((1, 1, 8))  # Shape: (batch_size, timesteps, features)
    current_input[0, 0, 0] = latest_data[0]  # Open
    current_input[0, 0, 1] = latest_data[1]  # High
    current_input[0, 0, 2] = latest_data[2]  # Low
    current_input[0, 0, 3] = latest_data[4]  # Adj Close
    current_input[0, 0, 4] = latest_data[5]  # Volume
    current_input[0, 0, 5] = 0  # Symbol index
    current_input[0, 0, 6] = np.random.uniform(0, 1)  # Positive sentiment (scaled)
    current_input[0, 0, 7] = np.random.uniform(0, 1)  # Negative sentiment (scaled)

    # Get min and max values for price scaling
    price_min = stock_data['Close'].min()
    price_max = stock_data['Close'].max()

    # Predict next 7 days
    for i in range(7):
        # Make prediction
        pred = model.predict(current_input, verbose=0)
        # Inverse transform the prediction (scale back to original price range)
        original_scale_pred = pred[0][0] * (price_max - price_min) + price_min
        predictions.append(original_scale_pred)

        # Prepare next day's input
        next_input = current_input.copy()
        next_input[0, 0, 0] = pred[0][0]  # Use predicted close as next day's opens
        next_input[0, 0, 1] = pred[0][0] * 1.01  # Estimate high as 1% above predicted
        next_input[0, 0, 2] = pred[0][0] * 0.99  # Estimate low as 1% below predicted
        next_input[0, 0, 3] = pred[0][0]  # Adj Close
        next_input[0, 0, 4] = current_input[0, 0, 4]  # Keep same volume
        next_input[0, 0, 5] = 0  # Symbol index
        next_input[0, 0, 6] = np.random.uniform(0, 1)  # New random positive sentiment
        next_input[0, 0, 7] = np.random.uniform(0, 1)  # New random negative sentiment

        current_input = next_input

        # Generate next date (skip weekends)
        next_date = end_date + timedelta(days=i+1)
        while next_date.weekday() > 4:  # Skip Saturday (5) and Sunday (6)
            next_date += timedelta(days=1)
        dates.append(next_date)

    # Create results DataFrame
    results_df = pd.DataFrame({
        'Date': dates,
        'Predicted_Close': predictions
    })

    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Close'], label='Historical Close')
    plt.plot(results_df['Date'], results_df['Predicted_Close'], label='Predicted Close', linestyle='--')
    plt.title(f'Stock Price Prediction for {symbol}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return results_df

# Example usage
if __name__ == "__main__":
    # Replace with your actual model path
    model_path = '/content/drive/MyDrive/model/model5/gru_model.keras'
    symbol = 'AAPL'  # Replace with your desired stock symbol

    predictions = predict_next_7_days(symbol, model_path)
    print("\nPredicted prices for next 7 days:")
    print(predictions)
```
